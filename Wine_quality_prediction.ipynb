```python
#importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data collection and analysis

```python
#loading the dataset in a pandas dataframe
wine_dataset=pd.read_csv("/content/wine quality.csv")
```

```python
#no. of rows and columns in dataset
wine_dataset.shape
```

```python
#printing first 5 rows of the dataset
wine_dataset.head()
```

```python
#checking for missing values
wine_dataset.isnull().sum()
```

```python
#data analysis and visualization
#statistical measures of the data
wine_dataset.describe()
```

```python
#no. of values for each quality
sns.catplot(x="quality",data=wine_dataset,kind="count")
```

```python
#Volatile acidity v/s Quality
plot=plt.figure(figsize=(5,5))
sns.barplot(x="quality",y="volatile acidity",data=wine_dataset)
```

```python
#Citric acid v/s Quality
plot=plt.figure(figsize=(5,5))
sns.barplot(x="quality",y="citric acid",data=wine_dataset)
```

```python
#data preprocessing
#seperate the data and lebels
x=wine_dataset.drop("quality",axis=1)
```

```python
print(x)
```

```python
#lebel binarization
y = wine_dataset["quality"].apply(lambda y_value: 1 if y_value >= 7 else 0)
```

```python
#Train and test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
#model training
#Random Forest Classifier
model=RandomForestClassifier()
model=model.fit(x_train,y_train)
```

```python
#model evaluation
#accuracy score
#accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("accuracy: ",test_data_accuracy)
```

```python
#Building a predictive system
input_data=(11.2,0.28,0.56,1.9,0.075,17,60,0.998,3.16,0.58,9.8)
```

```python
#changing the imput data to numpy array
input_data_as_numpy_array=np.asarray(input_data)
```

```python
#reshape the data as we are predicting the lebel for only one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
```

```python
prediction=model.predict(input_data_reshaped)
```

```python
print(prediction)
```

```python
if(prediction[0]==1):
  print("Good Quality Wine!")
else:
    print("Bad Quality Wine!")
```
